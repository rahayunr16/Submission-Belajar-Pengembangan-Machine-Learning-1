```python
from google_play_scraper import reviews, Sort
from google.colab import files
import pandas as pd
import time
```

```python
app_id = 'com.tokopedia.tkpd'

all_reviews = []

target_total = 15000

batch_size = 300

cursor = None
while len(all_reviews) < target_total:
    result, continuation_token = reviews(
        app_id,
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=batch_size,
        continuation_token=cursor
    )

    for r in result:
        all_reviews.append({
            'content': r['content'],
            'score': r['score']
        })

    cursor = continuation_token

    print(f"Jumlah data terkumpul: {len(all_reviews)}")

    time.sleep(1)
```

```python
df = pd.DataFrame(all_reviews)
df.head()
```

```python
df.to_csv('Tokopedia_Reviews_Scraping.csv', index=False)

files.download('Tokopedia_Reviews_Scraping.csv')
```
